```python
from os import listdir
from os.path import join, isfile, abspath, dirname

import sys
sys.path.insert(0, abspath(join(dirname("."), '..')))

import pickle as pkl
import csv

from random import sample, seed

from modules.helpers.utils import path_decompose, get_seed_for_graph
from modules.helpers.constants import PERCENTAGES, PERCENTAGE_DICT
```

```python
DS_PATH = path_decompose("H:\\Research\\NIGraphs\\withPasive")
SEED_PATH = path_decompose("H:\\Research\\NIGraphs\\seeds")

RUN_DEBUG = True
CURRENT_SEED = 2024
```

```python
seed(CURRENT_SEED)
```

```python
all_graphs = [f for f in listdir(DS_PATH) if isfile(join(DS_PATH, f))]
```

```python
for graph_name in all_graphs:
    graph_config = graph_name.replace(".pkl", "")

    # Read Graph Data
    graph_path = join(DS_PATH, graph_name)
    G = pkl.load(open(graph_path, 'rb'))

    if RUN_DEBUG:
        print(G)

    one_percent = int(len(G.nodes())/100)
    top_one_percent_nodes_by_degree = [int(x[0]) for x in sorted(
        G.degree, key=lambda x: x[1], reverse=True)[:one_percent]]

    if RUN_DEBUG:
        print(top_one_percent_nodes_by_degree)

    range_points = []
    for p in PERCENTAGES:
        range_points.append((p, int(PERCENTAGE_DICT[p] * one_percent)))

    # SampleForSeeds
    for (percentage, num_of_nodes_to_pick) in range_points:
        random_seeds = sample(
            top_one_percent_nodes_by_degree, num_of_nodes_to_pick)

        with open(get_seed_for_graph(graph_config, SEED_PATH, percentage), 'w') as myfile:
            wr = csv.writer(myfile)
            wr.writerow(random_seeds)
```
